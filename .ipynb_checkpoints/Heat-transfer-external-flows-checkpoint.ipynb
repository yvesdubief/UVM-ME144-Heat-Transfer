# Heat Transfer in External Flows

## Boundary Layer Flows
### Local vs. Global Quantities
Fig. 1 depicts a flow passing over a flat plate. The upstream flow has uniform mean temperature $T_\infty$ and velocity $U_\infty$. The plate is isothermal with a surface temperature $T_s>T_\infty$. Hereafter all thermodynamics properties in the flow, unless For this problem, there are two definitions of the Reynolds number, the local Reynolds number at a distance $x$ from the leading edge:
<p class='alert alert-danger'>
$$
Re_x\equiv\frac{U_\infty x}{\nu_f}
$$
</p>
and the global Reynolds number over the entire plate:
<p class='alert alert-danger'>
$$
Re_L\equiv\frac{U_\infty L}{\nu_f}
$$
</p>

Global and local quantities serve different purpose. Global or average quantities are the total drag or heat rate over the plate. Local quantities are necessary when the knowledge of the transition from laminar to turbulence is important, or making sure that the local temperature stays within the bounds of operation at all locations along the length of the plate. 

### Governing Factors in Convection Boundary Layers

Transport of heat or mass are interchangeable. In the first case, a plate could be hot (cold) to heat up (cool down) the flow, in the second a plate could release (absorb) molecular species, particles or aerosols through surface reactions into (from) the flow. The transport of heat or mass depends on 
* The nature of the flow (laminar vs turbulent) 
* The **molecular diffusion** of heat or mass in the fluid flowing over the plate, characterized by the thermal diffusivity $\alpha_f$ or the molecular diffusivity $D_{AB}$ of molecule/cells/aerosols/fine particle $A$ in fluid $B$
* The influence of gravity, also known as buoyancy or flow induced by gradients of gravity vs. inertial effects, or flow caused external forcing (fan, pump, pressure gradient, moving objective).

### Key Quantities in Convection Boundary Layers

When inertia dominates buoyancy for a flow over a flat plate, the flow is governed by:
* Wall friction: The basic measure of wall friction is **wall-shear stress**:
<p class='alert alert-danger'>
$$
\tau_s{x}=\mu_s\left.\frac{\partial u}{\partial y}\right\vert_{x,y=0}
$$
</p>
where $\mu_s$ is the dynamic viscosity at wall-temperature $T_s$. Derived the wall-shear stress, the **skin-friction velociy**
<p class='alert alert-danger'>
$$
\rho_s u_\tau^2=\tau_s
$$
</p>
This velocity characterizes the velocity scale $u_\tau$, length scale $l_v=\nu/u_\tau$ and time-scale $t_v=\nu/u_\tau^2$ of the dynamics of flow near the wall (in the viscous sublayer, buffer layer and lower part of the log-layer for turbulent flows). 
Lastly the coefficient of friction is the normalized wall-shear stress:
<p class='alert alert-danger'>
$$
C_f\equiv\frac{\tau_s}{\frac{1}{2}\rho U_\infty^2}
$$
</p>
* Wall-heat transfer: The key quantity is the heat flux given by Fourier's Law and Newton's Law of cooling:
<p class='alert alert-danger'>
$$
q_s''(x)=-k_f\left.\frac{\partial T}{\partial y}\right\vert_{x,y=0}=h(x)(T_s - T_\infty)
$$
</p>
where $h(x)$ is the local convection coefficient. The second important quantity is the Nusselt number
<p class='alert alert-danger'>
$$
Nu(x)=\frac{h(x)x}{k_f}=\frac{h(x)(T_s-T_\infty)}{-k_f(\partial T/\partial y\vert_{x,y=0})}
$$
</p>
* Upstream turbulence level: Beyond a certain level of turbulence, typically measured by the turbulent intensity $u'/U_\infty$, where $u'$ is a representative measure of the standard deviation or root-mean-square (RMS) of velocity fluctuations upstream of the leading edge
* Roughness of the plate's surface. 

### Kinematic and Thermal Boundary Layers
As you learned in Fluid Mechanics, a kinematic boundary layer grows downstream of the leading edge: this is the region of the flow where the gradient of the mean velocity profile is non-zero. In other words, the boundary layer is the region that smoothly connects the uniform flow region way from the wall to the wall where the velocity is zero.
The thickness of the kinematic boundary depends on the nature of the flow:
<p class='alert alert-danger'>
$$
\delta_u(x)=\begin{cases}

\end{cases}
$$
</p>
The critical quantity of convection is the Nusselt number, a non-dimensional number for the convection coefficient based on a reference length and the thermal conduction of the **fluid**. The convection coefficient is defined by considering that the wall heat flux from Newton's law with Fourier's law


$$
h(T_s-T_\infty)=-k_f\left.\frac{\partial T}{\partial y}\right\vert_{wall}
$$

The convection heat flux is **local**, for the flat plate it is a function of $x$, the distance from the leading edge. The calculation of the total heat rate is based on the **average** convection coefficient

<p class='alert alert-danger'>
$$
\bar{h}=\frac{1}{L}\int_0^Lh(x)dx
$$
</p>

The Nusselt number is defined as
<p class='alert alert-danger'>
$$
Nu(x)=\frac{h(x)x}{k_f}\text{ and } \overline{Nu}=\frac{\bar{h}L}{k_f}
$$
</p>
and is typically a function of the Reynolds number and Prandtl number for external forced convection.

<p class='alert alert-danger'>
$$
Re_x=\frac{U_\infty x}{\nu_f},\;Pr=\frac{\nu_f}{\alpha_f}
$$
</p>
The viscosity $\nu_f$, thermal diffusivity $\alpha_f$ and thermal conductivity $k_f$ are thermodynamic properties of the <b>fluid</b> generally calculated at the fluid temperature

<p class='alert alert-danger'>
$$
T_f=\frac{T_s+T_\infty}{2}
$$
</p>
Alternative methods include the determination of the thermodynamics at the free stream temperature $T_\infty$ and a correction factor base $Pr_\infty/Pr_s$ as will be seen in some Nusselt number correlations. 

The end goal is to identify and use a Nusselt number correlation for the flow of interest. For forced external convection, this correlation takes the form:

<p class='alert alert-danger'>
$$
Nu_x=F(Re_x,Pr)\text{ or }\overline{Nu}_L=G(Re_L,Pr)
$$
</p>


## Flow over a plate

For this flow (shown in Fig. 1), one must figure out the following:
- The free stream velocity and temperature conditions and relevant thermodynamic properties at the appropriate temperatures (either $T_f$, or $T_\infty$ and $T_s$.
- Is the plate isothermal or constant flux? 
- Is there an unheated starting length?
- is the surface smooth or rough, i.e. can the flow be considered turbulent over the entire plate?
- Is the desired outcome local heat flux or average heat rate over the plate?

### Creating a boundary layer object

In the python heat transfer library, all correlations for a flat plate are coded under one object `FlatPlate` in the library `HT_external_convection`
```Python
from Libraries import HT_external_convection as extconv
bl = extconv.FlatPlate(regime,thermal_bc,U_infty,nu_f,alpha_f,L,xi,Re_xc)
```
where
- `regime` is `'laminar'`, `'turbulent'` or `'mixed'`. The first two regimes assume that the regime is either laminar or turbulent over the *entire* length of the plate. If you suspect that the flow may transition to turbulent, i.e. $Re_L$ based on the length of the plate is greater than $Re_{x,c}$, use `'mixed'`. Tip: using `'mixed'` when $Re_L\leq Re_{x,c}$ is the same as using `'laminar'`
- `thermal_bc` is either `'isothermal'` or `'heat flux'` for constant wall temperature $T_s$ or constant wall heat flux $q''_s$
- `U_infty` is the free stream velocity
- `nu_f` is the **kinematic** viscosity ($\mathrm{m^2/s}$) at $T_f$
- `alpha_f` is the thermal diffusivity at ($\mathrm{m^2/s}$)at $T_f$
- `L` is the length of the plate
- `xi` is the unheated starting length from the leading edge (**default $\xi=0$**)
- `Re_xc` is the critical Reynolds number at which the flow regime transitions from laminar to turbulent. (**default $Re_{x,c}=5\times10^5$**)

Note 1: Object parameters that have an assigned default value **do not need to be specify in the call of the object if there default values apply to your problem**

Note 2: As for other functions/objects, the order of parameters **matters**.

#### Example: mixed regime, constant heat flux, no unheated starting length:
```Python
from Libraries import HT_external_convection as extconv
mybl = extconv.Flatplate('mixed','heat flux',U_inf,nu_f,alpha_f,L)
```

#### Example: mixed, isothermal with unheated starting length and $Re_{x,c}=10^4$:
```Python
myxi = L/4
from Libraries import HT_external_convection as extconv
mybl = extconv.Flatplate('mixed','isothermal',U_inf,nu_f,alpha_f,L,xi=myxi,Re_xc=1e4)
```
#### Example: turbulent, isothermal with unheated starting length:
```Python
myxi = L/4
from Libraries import HT_external_convection as extconv
mybl = extconv.Flatplate('turbulent','isothermal',U_inf,nu_f,alpha_f,L,xi=myxi)
```
The first and third examples assume default values for the unspecified parameters.

### Computing the local or average Nusselt number
Now that the object is created, the Nusselt number (and friction coefficient) can be computed for most liquid and gases (the liquid metal correlation has not been coded yet). For example the average Nusselt number correlations for the laminar, turbulent and mixed regimes are, respectively,:
$$
\overline{Nu}_L=0.664Re_L^{1/2}Pr^{1/3},\;Pr\geq 0.6
$$
$$
\overline{Nu}_L=0.037Re_L^{4/5}Pr^{1/3},\;Pr\geq 0.6
$$
$$
\overline{Nu}_L=\left[(0.037(Re_L^{4/5}-Re_{x,c}^{4/5})+0.664Re_{x,c}^{1/2}\right]Pr^{1/3},\;Pr\in[0.6,60],\;Re_L\in[Re_{x,c},10^8]
$$

#### Example: local quantities
First, run the function `local`, where `x` is the numerical value where the local quantities are to be determined. The code below shows how to get $Nu_x$, $Cf_x$, $Re_x$, $\delta(x)$ (velocity boundary layer thickness), $\delta_T(x)$ (thermal boundary layer thickness)
```Python
from Libraries import HT_external_convection as extconv
mybl = extconv.Flatplate('turbulent','isothermal',U_inf,nu_f,alpha_f,L,xi=myxi)
mybl.local(x)
Nux = mybl.Nu_x
Cfx = mybl.Cf_x
Rex = mybl.Re_x
deltax = mybl.delta_x
deltaTx = mybl.delta_Tx
```

#### Example: Average quantities
Here we use the function `average` for a length $L$
```Python
from Libraries import HT_external_convection as extconv
mybl = extconv.Flatplate('turbulent','isothermal',U_inf,nu_f,alpha_f,L,xi=myxi)
mybl.average(L)
NuLave = mybl.Nu_ave
CfLave = mybl.Cf_ave
```

## Now what?

Now that you know how to calculate the local and average Nusselt number, you can calculate:
- the local heat flux:
$$
q''(x)=h(x)(T_s-T_\infty)=\frac{Nu_xk_f}{x}(T_s-T_\infty)
$$
- the total heat transfer rate (per unit width of the plate):
$$
q' = \bar{h}_LL(T_s-T_\infty)=\frac{\overline{Nu}_Lk_f}{L}L(T_s-T_\infty)
$$

# Design of a low temperature ablation experimental facility

During atmopheric entry, the heat shield of a space vehicle experiences tremendous heat load and aggressive chemical reactions. On some materials, the formation of roughness patterns is observed. These patterns are caused by the local ablation of the matieral. Roughness may cause extra heat flux through increased turbulence that could lead to the failure of the heat shield. The mechanisms by which such ablation or erosion patterns are formed are poorly understood, leaving engineers with few if any ability to predict the outcome of the heat shield. 

Here we consider a simple system made out of a plate of wax of thickness $H=0.05\text{ m}$ in order to study this system in a lab without worrying about high-temperatures.

The local ablation velocity of wax $V_a(x)$ (i.e. the removal rate of solid wax) is derived from a conservation of energy:

$$
\rho_{wax}h_lV_a(x) =\begin{cases}\max\left(q''_\text{conv}-q''_\text{cond},0\right)&\mathrm{if}\; T_s(x)\geq T_m\\
0& \text{otherwise}\end{cases}
$$

where $\rho_{wax}=900\,\mathrm{kg}/\mathrm{m}^3$ is the wax density, $h_l= 200\;\mathrm{kJ}/\mathrm{kg}$ is the latent heat of wax and $q''_\text{conv}$ and $q''_\text{cond}$ are the *local* heat fluxes by convection and conduction, respectively. $T_s(x)$ an $T_m$ are the *local* surface temperature, and the surface temperature, respectively.

The goal is to study the formation of erosion patterns on the surface of the wax plate, thus we are interested in maximumizing the ablation velocity. We will consider two fluids, air and water. 

For both case, the bottom of the wax plate is maintained at constant temperature $T_\text{cool}=20^\circ\mathrm{C}$. 
In the following, you are asked to find the initial ablation velocity assuming that (a) the plate has not started to melt yet and (b) the flow-exposed surface is at melting temperature $T_\text{melt}=50^\circ\mathrm{C}$. The thermal conductivity of wax is $k_{wax}=0.25\mathrm{W}/\mathrm{m.K}$ and radiation heat transfer is neglected.

1) The upstream air flow velocity is $10 \,\mathrm{m/s}$ and its upstream temperature is $150^\circ\mathrm{C}$. Plot the evolution of the initial ablation velocity in $\mathrm{mm/s}$ and/or give its values for $x=0.01,0.1,0.2,0.5,1.0\mathrm{m}$. The leading edge is a very thin, adiabatic rounded edge.

The heat flux by conduction within the plate is uniform (before any melt occurs):
$$
q''_{cond}=\frac{k_{wax}}{t}(T_{melt}-T_{bot})
$$
where t is the thickness of the wax plate.

For the convection heat flux, we use the Nusselt number for isothermal plate. The plate is assumed to be smooth. The thermodynamics properties of the fluid are estimated at the fluid temperature

$$
T_f=\frac{T_\infty+T_{melt}}{2}
$$

The convection heat flux is
$$
q''_{conv}(x)=Nu(x)\frac{k_f}{x}(T_\infty-T_{melt})
$$
where $Nu(x)$ is obtained from the heat transfer libraries.

Therefore the local ablation velocity is 

$$
V_a(x)=\frac{1}{\rho_{wax}h_l}(q''_{conv}(x)-q''_{cond})
$$

Note: The thermal circuit below illustrates the cooling effect of ablation: the energy of convection in excess of the energy evacuated by cooling the bottom surface of the wax goes into removal of material by phase change:
$$
q''_{conv}=\rho_{wax}h_lV_a(x)+q''_cond
$$
When integrated over the entire surface of dimensions $L\times W$ (L: length, W: span) and acknowledging that the ablation velocity is the time-derivative of the local height of the slab,
$$
V_a(x,t)=\frac{-\partial H(x,t)}{\partial t},
$$
this equation becomes
$$
q_{conv} = h_l\int_0^L\rho_{wax}\left(-\frac{\partial H(x,t)}{\partial t}\right)Wdx =-h_L\frac{\partial}{\partial t}\int_0^L\rho_{wax}H(x,t)Wdx=-h_L\frac{\partial M_{wax}(t)}{\partial t}
$$
where $M_{wax}(t)$ is the mass of the wax slab

````python
import schemdraw as schem
import schemdraw.elements as elm
from Libraries import HT_thermal_resistance as res

R = []
R.append(res.Resistance("$R''_{conv}$",'W/m^2'))
R.append(res.Resistance("$R''_{cond}$",'W/m^2'))

with schem.Drawing() as d:
    d += elm.Dot().label("$T_\infty$")
    d += elm.Resistor().label(R[0].name)
    d += elm.Dot().label("$T_s$")
    d += elm.Resistor().label(R[1].name)
    d += elm.Dot().label("$T_{cool}$")
    d += elm.Line(arrow='->').at((-2.25,0)).length(2).label("$q''_{conv}(x)$")
    d += elm.Line(arrow='->').down().at((3,-0.25)).length(2).label(r"$\rho_{wax}h_lV_a(x)$")
    d += elm.Line(arrow='->').right().at((6.25,0)).length(2).label("$q''_{cond}$")
    
````

````python
T_inf = 150. #C
U_inf = 10 #m/s
T_melt = 50 #C
T_bot = 20 #C
rho_wax =900 #kg/m^3
h_l = 200e3 #J/kg
k_wax = 0.25 #W/m.K
H = 0.05 #m

# thermodynamics of the fluid
from Libraries import thermodynamics as thermo
T_f = (T_inf + T_melt)/2
air = thermo.Fluid('air',T_f,"C")

# Nusselt number as a function of x Nu(x)=f(Re_x,Pr)
x = np.array([0.01, 0.1, 0.2, 0.5, 1.0])
x = np.logspace(-4,0,401)
Rex = U_inf*x/air.nu
print("Re_x at the end of the plate: %.1e" %Rex[-1])

from Libraries import HT_external_convection as extconv

bl = extconv.FlatPlate('laminar','isothermal',U_inf,air.nu,air.alpha,x[-1])
bl.local(0.01)
print(bl.Nu_x)
Nux = np.zeros_like(x)
for i in range(len(Nux)):
    bl.local(x[i])
    Nux[i] = bl.Nu_x
    
qpp_conv = Nux*air.k/x*(T_inf - T_melt)
plt.figure(figsize=(4,3),dpi=100)
plt.plot(x,qpp_conv)
plt.title("Convection heat flux")
plt.ylabel("$q''_{conv}$")
plt.xlabel("$x$")
plt.show()
````

````python
# conduction heat flux inside wax plate
qpp_cond = k_wax/H*(T_melt - T_bot)

Vax = (qpp_conv - qpp_cond)/(rho_wax*h_l)
plt.figure(figsize=(4,3),dpi=100)
plt.plot(x,Vax*1e3)
Vax_air = np.copy(Vax)
plt.title("Ablation velocity at time $t=0$")
plt.ylabel("Ablation velocity (mm/s)")
plt.xlabel("$x$")
plt.show()
````

Plot the evolution of the initial ablation velocity in $\text{mm/s}$ and/or give its values for $x=0.01,0.1,0.2,0.5,1.0\mathrm{m}$.
````python
T_inf = 70. #C
U_inf = 10 #m/s
T_melt = 50 #C
T_bot = 20 #C
rho_wax =900 #kg/m^3
h_l = 200e3 #J/kg
k_wax = 0.25 #W/m.K
H = 0.05 #m

# thermodynamics of the fluid
from Libraries import thermodynamics as thermo
T_f = (T_inf + T_melt)/2
water = thermo.Fluid('water',T_f,"C")

# Nusselt number as a function of x Nu(x)=f(Re_x,Pr)
Rex = U_inf*x/water.nu
print("Re_x at the end of the plate: %.1e" %Rex[-1])

from Libraries import HT_external_convection as extconv

bl = extconv.FlatPlate('mixed','isothermal',U_inf,water.nu,water.alpha,x[-1])
bl.local(0.01)
print(bl.Nu_x)
Nux = np.zeros_like(x)
for i in range(len(Nux)):
    bl.local(x[i])
    Nux[i] = bl.Nu_x
    
qpp_conv = Nux*water.k/x*(T_inf - T_melt)
plt.figure(figsize=(4,3),dpi=100)
plt.semilogx(x,qpp_conv)
plt.title("Convection heat flux")
plt.ylabel("$q''_{conv}$")
plt.xlabel("$x$")
plt.show()
````

````python
# conduction heat flux inside wax plate
qpp_cond = k_wax/H*(T_melt - T_bot)

Vax = (qpp_conv - qpp_cond)/(rho_wax*h_l)
plt.figure(figsize=(4,3),dpi=100)
plt.semilogx(x,Vax*1e3)
plt.title("Ablation velocity at time $t=0$")
plt.ylabel("Ablation velocity (mm/s)")
plt.xlabel("$x$")
plt.show()
````
